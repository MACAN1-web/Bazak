```python
import numpy as np
import matplotlib.pyplot as plt



def f(x):
    return 2 * np.log10(x + 7) - 5 * np.sin(x)  


x_vals = np.linspace(-6, 5, 1000)  
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = 2*lg(x+7) - 5*sin(x)')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.axvline(-7, color='red', linestyle='--', linewidth=0.8, label='x = -7 (граница ОДЗ)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Графическое отделение корней (вариант 11)')
plt.grid(True)
plt.legend()
plt.show()


intervals = [(-2, 0), (1, 3)]  


def find_sign_intervals(a, b, steps=100):
    x_vals = np.linspace(a, b, steps)
    intervals = []
    for i in range(len(x_vals) - 1):
        if f(x_vals[i]) * f(x_vals[i + 1]) <= 0:
            intervals.append((x_vals[i], x_vals[i + 1]))
    return intervals


all_intervals = find_sign_intervals(-6, 5, 500)
print("Интервалы, содержащие корни (по знаку функции):")
for interval in all_intervals:
    print(f"  {interval}")


chosen_interval = (1.5, 2.5)
print(f"\nВыбранный интервал для уточнения корня: {chosen_interval}")
```
