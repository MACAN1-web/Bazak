```python
import numpy as np


def f(x):
    return 2 * np.log10(x + 7) - 5 * np.sin(x)


print("Уравнение: 2·lg(x+7) - 5·sin(x) = 0")
print("Итерационная форма: x = x + 0.1·f(x)")
print(f"{'№':<4} {'x_k':<14} {'x_{k+1}':<14} {'|x_{k+1}-x_k|':<18} {'f(x_k)':<12}")
print("-" * 70)


x = 2.000977
lambd = 0.1
k = 0

while True:
    f_val = f(x)
    x_next = x + lambd * f_val
    diff = abs(x_next - x)
    
    print(f"{k:<4} {x:<14.8f} {x_next:<14.8f} {diff:<18.12f} {f_val:<12.8f}")
    
    if diff < 1e-6:
        break
        
    x = x_next
    k += 1

print("-" * 70)
print(f"\nКорень уравнения: x = {x_next:.8f}")
print(f"Проверка: f({x_next:.8f}) = {f(x_next):.2e}")
```
