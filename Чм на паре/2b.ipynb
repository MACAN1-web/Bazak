```python
import numpy as np


def f(x):
    return 2 * np.log10(x + 7) - 5 * np.sin(x)


print("=" * 70)
print("ЗАДАНИЕ 2а. МЕТОД ПОЛОВИННОГО ДЕЛЕНИЯ (РУЧНОЙ РАСЧЕТ)")
print("=" * 70)
print("Уравнение: 2·lg(x+7) – 5·sin x = 0")
print("Интервал: [1.5, 2.5]  (определен графически)")
print("Точность: ε = 0.001")
print("-" * 70)


a, b = 1.5, 2.5
fa = f(a)
fb = f(b)
n = 0

print(f"{'№':<3} {'a':<8} {'b':<8} {'c':<8} {'f(a)':<10} {'f(b)':<10} {'f(c)':<10} {'|b-a|/2':<10}")
print("-" * 70)

while (b - a) / 2 > 0.001:
    c = (a + b) / 2
    fc = f(c)
    print(f"{n:<3} {a:<8.4f} {b:<8.4f} {c:<8.4f} {fa:<10.4f} {fb:<10.4f} {fc:<10.4f} {(b-a)/2:<10.6f}")
    
    if fa * fc < 0:
        b = c
        fb = fc
    else:
        a = c
        fa = fc
    n += 1


c = (a + b) / 2
fc = f(c)
print(f"{n:<3} {a:<8.4f} {b:<8.4f} {c:<8.4f} {fa:<10.4f} {fb:<10.4f} {fc:<10.4f} {(b-a)/2:<10.6f}")
print("-" * 70)

print(f"\n Корень (с точностью 0.001): x = {c:.6f}")
print(f"   Проверка: f({c:.6f}) = {fc:.6f}")

```
